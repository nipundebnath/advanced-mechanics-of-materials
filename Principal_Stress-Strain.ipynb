# **PRINCIPAL STRESSES** & **PRINCIPAL AXES**

```python
import numpy as np
import matplotlib.pyplot as plt
from utils.plot_tensor import plot_tensor_element

%matplotlib inline
```

## **Stress Tensor**

The stress tensor $[\sigma]$ is given bellow:

```python
stress_tensor = np.array([
    [-10,  15,  70],
    [ 15,  30, -40],
    [ 70, -40,   0],
], 
    dtype='float32' )
```

## **Calculation**

Given a principal direction $\vec{n}$ and principal stress $\tilde{\sigma}$ in this direction, we know:

$$
\left( [\sigma] - \tilde{\sigma}[I] \right) \vec{n} = 0 \
$$

It is en eigenvalue problem where $\tilde{\sigma}$ and $\vec{n}$ are eigenvalue and eigenvector, respectively, for the square matrix $[\sigma]$.

We can use `numpy.linalg.eig()` finction to directly calculate the eigenvalues and eigenvectors.

```python
### Eigen value and Eigen vector
principal_stress, principal_axes = np.linalg.eig( stress_tensor )

### Converts to a 3x3 tensor
principal_stress_tensor = np.eye(3) * principal_stress
```

```python
### Print the values
np.set_printoptions(precision=2, suppress=True)

print('Principal stress tensor:')
print(principal_stress_tensor)
print()

print('Principal axes:')
print(principal_axes)
```

## **Visualization**

```python
fig = plt.figure( figsize=(12,6) )

### Original stresses
ax = fig.add_subplot(121, projection='3d')
ax = plot_tensor_element(ax, stress_tensor, axes=None, max_value=np.max(principal_stress))
plt.title( 'Original Stress Tensor', y=-0.01)

### Principal stresses
ax = fig.add_subplot(122, projection='3d')
ax = plot_tensor_element(ax, principal_stress_tensor, axes=principal_axes, max_value=np.max(principal_stress))
plt.title( 'Principal Stress Tensor', y=-0.01)

plt.show()
```
